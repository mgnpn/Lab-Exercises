```python
PATH = Service("C:\Program Files (x86)\chromedriver_win32\chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://data.gov/")

import csv
csv_file = open("climatechange_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Description"])

search = driver.find_element(By.ID, "search-header")
search.clear()
search.send_keys("climate change")
search.send_keys(Keys.RETURN)

try:
    main = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.CLASS_NAME, "module-content")) #locates main ID
    )
    #print(main.text)
    
    articles = main.find_elements(By.CLASS_NAME, "dataset-content") #locates article tag
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "dataset-heading") #locates entry-summary class 
        title = header.find_element(By.TAG_NAME, "a").text
        print(title)
        
        notes = i.find_element(By.CLASS_NAME, "notes")
        desc = notes.find_element(By.TAG_NAME, "div").text
        print(desc)
        
        csv_writer.writerow([title, desc])
finally:
        driver.quit()
        
csv_file.close()
```
