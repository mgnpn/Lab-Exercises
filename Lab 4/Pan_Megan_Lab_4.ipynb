```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

# IMDB

```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")

#print(soup)
```

```python
section = soup.find("div", class_="lister-item mode-detail")
movie = section.h3.a.text
release_date = section.find("span", class_="lister-item-year text-muted unbold").text.strip("()")
summary = section.find("p", class_ = "").text
#print(section)
print(movie)
print(release_date)
print(summary)
```

```python
csv_file = open("imdb_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Movie Title", "Release Date", "Summary"])

for element in soup.find_all("div", class_="lister-item mode-detail"):
    
    movie = element.h3.a.text
    release_date = element.find("span", class_="lister-item-year text-muted unbold").text.strip("()")
    summary = element.find("p", class_ = "").text.strip("\n")
    
    csv_writer.writerow([movie, release_date, summary])
    
csv_file.close()
```

# Library of Congress

```python
source = requests.get("https://www.loc.gov/search/?q=cats&sp=1").text
soup = BeautifulSoup(source, "lxml")

#print(soup)
```

```python
def get_text(element):
    if element is not None:
        return element.text
    return "N/A"
```

```python
section = soup.find("div", class_="item-description")
title = section.a.text.strip()
description = get_text(section.find("span", class_="item-description-abstract")).strip()
link = section.a["href"]
#print(section)
print(title)
print(description)
print(link)
```

```python
new_csv = open("loc_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_csv)
csv_writer.writerow(["Title", "Description", "Link"])

page = 1

while page != 6:
    cats = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup(cats, "lxml")
    
    for element in soup.find_all("div", class_="item-description"):

        title = element.a.text.strip()
        description = get_text(element.find("span", class_="item-description-abstract")).strip()
        link = element.a["href"]
    
        csv_writer.writerow([title,description,link])
   
    page += 1
    
csv_file.close()
```
